# Task 1: Setting up an API key and .env file (1p)

```python
from dotenv import load_dotenv
from datetime import date
import os
import requests
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
# Load all variables from .env into the environment
load_dotenv()
api_key = os.environ.get("MARKETSTACK_API_KEY")

if api_key:
    print(f"Key loaded: {api_key[:5]}...")
else:
    print("ERROR: MARKETSTACK_API_KEY not found. Check your .env file.")
```

---
# Task 2: Retrieving data from MarketStack (1p)

```python
#a)
tickers = [
    'META',
    'AAPL',
    'AMZN',
    'NFLX',
    'GOOG',
]

```

```python
url = f"http://api.marketstack.com/v1/eod?access_key={api_key}&symbols={','.join(tickers)}&date_from=2026-03-01&date_to=2026-03-31"
response = requests.get(url)
```

```python
#b)
data = response.json()
stock = data['data']
df = pd.DataFrame(data=stock)
df.head()
```
