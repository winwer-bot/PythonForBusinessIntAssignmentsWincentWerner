# Task 1: A list of dictionaries
---

```python

transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

#a)
print("a)")
# Create new dictionary and add items that match critera to it
filtered = []
for dictionary in transactions: 
    if dictionary["price"] > 0 and dictionary["quantity"] >= 0:
        filtered.append(dictionary)

#b)
print("b)")
for transaction in filtered:
    transaction["total"] = transaction["price"] * transaction["quantity"]

#c)
print("c)")
# Creating three lists and adding their respective transactions to them
EU = []
US = []
APAC = []
for transaction in filtered:
    if transaction["region"] == "EU":
        EU.append(transaction)
    elif transaction["region"] == "US":
        US.append(transaction)
    else:
        APAC.append(transaction)

totals = {}

regiontotal = 0
for transaction in EU: 
    regiontotal += transaction["total"]
totals["EU"] = regiontotal

regiontotal = 0
for transaction in US: 
    regiontotal += transaction["total"]
totals["US"] = regiontotal

regiontotal = 0
for transaction in APAC: 
    regiontotal += transaction["total"]
totals["APAC"] = regiontotal


#d)
print("d)")
largest = 0
final_region=""

for region in totals:
    if totals[region] > largest:
        largest = totals[region]
        final_region = region

print(f"the largest region by revenue is {final_region}, and its total revenue is {largest}.")
```

# Task 2: creating a function
---

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(summarise):
    dict = {}
    dict["total"] = sum(summarise)
    dict["average"] = round(dict["total"]/len(summarise), 2)
    dict["best_day"] = summarise.index(max(summarise))+1
    dict["worst_day"] = summarise.index(min(summarise))+1

    count = 0
    for day in summarise:
        if day > dict["average"]:
            count += 1
    dict["above_average_count"] = count
    
    return dict
    
print(summarise_sales(sales))
```

# Task 3: Loops
---

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

# a)
print("a)")
passing = []

for student in students:
    if student["grade"] >= 50:
        passing.append(student["name"])

print(f"Passing students: {passing}")

#b)
print("b)")
for student in students:
    message = "Excellent"
    if student["grade"] < 80:
         message = "Pass"
    if student["grade"] < 50:
        message = "Fail"
    name = student["name"]
    print(f"{name}: {message}.")

#c)
print("c)")
count = 0
for student in students:
    if student["attendance"] < 0.8:
        count+=1
print(f"Number of students with attendace over 0.80: {count}.")

#d)
print("d)")
name = ""
best_score = 0
for student in students:
    if student["grade"] > best_score:
        best_score = student["grade"]
        name = student["name"]
print(f"The best student is {name}, with a score of {best_score}.")

#e)
print("e)")
#https://stackoverflow.com/questions/5320871/how-to-find-the-min-max-value-of-a-common-key-in-a-list-of-dicts

best_grade = students.index(max(students, key=lambda student: student['grade'])) #The key argument uses a lambda function to find which dictionary entry to use to compare using max() function. My first time stumbling upon lambda functions, very nifty
print(f"The best student is {students[best_grade]['name']}, with a score of {students[best_grade]['grade']}.")
```

Task 4: Class and methods
---

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

#a) 
    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings:
            self.holdings[ticker]["shares"] += shares
            self.holdings[ticker]["price_per_share"] = price_per_share
        else: self.holdings[ticker] = {"shares": shares, "price_per_share": price_per_share}

#b) 
    def total_value(self):
        total = 0
        for share in self.holdings.values():
           total += (share["shares"])*(share["price_per_share"])
        return total

#c) 
    def most_valuable(self):
        return max(self.holdings, key=lambda ticker: self.holdings[ticker]["shares"] * self.holdings[ticker]["price_per_share"])

#d)
    def __str__(self):
        return f"Portfolio owner: {self.owner}, Number of assets: {len(self.holdings)}, Total value: {self.total_value()}"

portfolio = Portfolio("Wincent")
portfolio.add_asset("HEL: NOKIA", 200, 1.50)
portfolio.add_asset("INTC", 5, 47.20)
portfolio.add_asset("MSFT", 1, 350)
print(f"The total value of the portfolio is {portfolio.total_value()}.")
print(f"The most valuable holding value of the portfolio is{portfolio.most_valuable()}.")
print(portfolio.__str__())
```

```python
#https://docs.python.org/3/library/datetime.html#date-objects

from datetime import datetime

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

dates = []
for order in orders:
    dates.append(dt.strptime(order[0], "%Y-%m-%d").date())

earliest = min(dates)
latest = max(dates)
difference = latest-earliest

print(f"earliest: {earliest}")
print(f"latest: {latest}")
print(f"difference: {difference}")
```
